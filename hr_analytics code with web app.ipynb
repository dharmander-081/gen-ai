```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data
df = pd.DataFrame({
    "Salary": [50000, 60000, 55000, 70000],
    "Base_pay": [30000, 35000, 32000, 40000],
    "Bonus": [5000, 7000, 6000, 8000],
    "Age": [25, 28, 26, 30]
})

# Set style
sns.set(style="darkgrid")

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot boxplots
sns.boxplot(data=df, x="Salary", color="lime", ax=axs[0, 0])
sns.boxplot(data=df, x="Base_pay", color="olive", ax=axs[0, 1])
sns.boxplot(data=df, x="Bonus", color="gold", ax=axs[1, 0])
sns.boxplot(data=df, x="Age", color="pink", ax=axs[1, 1])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")

# Convert columns to numeric if needed
for col in ["low", "Total_Sales", "Volume", "Months"]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot boxplots
sns.boxplot(data=df, x="low", color="red", ax=axs[0,0])
sns.boxplot(data=df, x="Total_Sales", color="navy", ax=axs[0,1])
sns.boxplot(data=df, x="Volume", color="yellow", ax=axs[1,0])
sns.boxplot(data=df, x="Months", color="skyblue", ax=axs[1,1])

plt.tight_layout()
plt.show()
```
